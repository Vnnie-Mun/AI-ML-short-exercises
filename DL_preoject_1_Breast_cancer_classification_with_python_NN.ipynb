importing the dependancies


```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Data collection and pre-processing

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(  breast_cancer_dataset)
```

```python
#loading the data to a pandas DataFrame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
#print the first 5 rows of the dataframe
data_frame.head()
```

```python
#adding the 'target' cloumn to the DataFrame
data_frame['label'] = breast_cancer_dataset.target
```

Separating the featurees and targets

```python
X = data_frame.drop(columns='label',axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

Building a neural network

```python
#importing tensorflow and Kreas

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
    ])
```

```python
#compiling the Neural Network

model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
#training the neural network

history= model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

Visualizing accuracy loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy score of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```

Building the predictive system

```python
# Use the first data point from X_test as input data
input_data = X_test_std[0]

#change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape thenumpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardizing the input data
# input_data_std = scaler.transform(input_data_reshaped) # Already standardized

prediction = model.predict(input_data_reshaped)
print(prediction)

prediction_labels = [np.argmax(prediction)]
print(prediction_labels)

if (prediction_labels[0] == 0):
  print('The breast cancer is Benign')
else:
  print('The breast cancer is Malignant')
```
